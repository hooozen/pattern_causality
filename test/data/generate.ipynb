$$
X(t+1) = 0.2X(t)[1-X(t)/K_X + \varphi(p)0.1Y(t)/K_X], X \in \mathbb R^+, K_X\in \mathbb R^+. \\

Y(t+1) = 0.2Y(t)[1-Y(t)/K_Y + \varphi(p)0.2X(t)/K_Y], Y \in \mathbb R^+, K_Y\in \mathbb R^+.
$$

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def phi_p():
    p = np.random.random(1)[0]
    if p < 0.3:
        return 0
    return 1
```

```python
L = 200
K_X = 100
K_Y = 100

x = np.empty(200)
y = np.empty(200)
x[0] = 10
y[0] = 5
for t in np.arange(L-1):
    x[t + 1] = x[t] + 0.2 * x[t] * \
        (1 - x[t] / K_X + phi_p() * 0.1 * y[t] / K_X)
    y[t + 1] = y[t] + 0.2 * y[t] * \
        (1 - y[t] / K_Y + phi_p() * 0.2 * x[t] / K_Y)
```

```python
fig, axe = plt.subplots()
axe.plot(np.arange(0, L), x, label='x')
axe.plot(np.arange(0, L), y, label='y')
axe.set_ylim(50, None)
axe.legend()

plt.show()
```

```python
np.savetxt('mutualism.csv',  np.vstack((x, y)), delimiter=",")
```

```python
L = 200
K_X = 100
K_Y = 100

x = np.empty(200)
y = np.empty(200)
x[0] = 50
y[0] = 50
for t in np.arange(L-1):
    x[t + 1] = x[t] + 0.7 * x[t] * \
        (1 - x[t] / K_X - phi_p() * 0.2 * y[t] / K_X)
    y[t + 1] = y[t] + 0.7 * y[t] * \
        (1 - y[t] / K_Y - phi_p() * 0.3 * x[t] / K_Y)
```

```python
fig, axe = plt.subplots()
axe.plot(np.arange(0, L), x)
axe.plot(np.arange(0, L), y)
axe.set_ylim(50, None)

plt.show()
```
