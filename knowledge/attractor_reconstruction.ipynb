```python
from scipy.integrate import odeint
import numpy as np
from matplotlib import pyplot as plt
```

```python
def lorenze(w, t, p, r, b):
    x, y, z = w
    return [p * (y - x), x * (r - z) - y, x * y - b * z]


t = np.arange(0, 30, 0.01)
M = odeint(lorenze, [0, 1.00, 0], t, args=(10, 28, 8 / 3))
```

```python
x, y, z = M.T
ax = plt.figure(figsize=(8, 8)).add_subplot(projection='3d')
ax.plot(x,y,z)
plt.show()
```

```python
fig, axe = plt.subplots(1, 3, figsize=(12, 4))
axe[0].plot(x)
axe[1].plot(y)
axe[2].plot(z)
```

```python
L = x.size
mx = np.array([x[2:], x[1: L - 1], x[: L - 2]])
my = np.array([y[2:], y[1: L - 1], y[: L - 2]])

x, y, z = mx
ax = plt.figure(figsize=(8, 8)).add_subplot(projection='3d')
ax.plot(x, y, z)
x, y, z = my
ax = plt.figure(figsize=(8, 8)).add_subplot(projection='3d')
ax.plot(x, y, z)

plt.show()
```
